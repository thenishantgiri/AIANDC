```python
# TREE 1
 
OPEN=[]    
CLOSE=[]   
 
s=0  
 
final='G' 
 
print("Please enter the elements of the tree\n")   
 
print("Please enter first element (root node)\n") 
root = input()
OPEN.append(root)    
 
if root==final:      
  print("Goal Node Found!") 
  print("The path is", root)
  s=1                       
else:
  OPEN.pop()                 
  CLOSE.append(root)       
  
 
currentnode = root        
 
ans='y' 
while s==0 and (ans=='y'or ans=='Y'):       
  print ("Please enter the LEFT child of", currentnode) 
  left=input()
  print ("Please enter the heuristic value of", left) 
  HL=int(input())
  print ("Please enter the RIGHT child of", currentnode)
  right=input()
  print ("Please enter the heuristic value of", right) 
  HR=int(input())
  
  if HL<HR:             
    CLOSE.append(left)  
    OPEN.append(right) 
    currentnode=left    
    if left==final:    
      print("Goal Node Found!") 
      print("The path is")
      print(CLOSE)  
      s=1           
      break
  else:
    CLOSE.append(right) 
    OPEN.append(left)   
    currentnode=right    
    if right==final:    
      print("Goal Node Found!")  
      print("The path is")
      print(CLOSE)      
      s=1               
      break
 
  print("OPEN List =", OPEN)        
  print("CLOSE List =", CLOSE)      
 
  print("Do you want to continue(y/n)?")    
  ans=input()
 
if s==0:
   print("Goal not found !")
```
