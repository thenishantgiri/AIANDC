```python
#TREE 1
GAMETREE=['A','B','C','D','E','F','G', 'H', 'I', 'J', 'K', 'L', 'M', 'N','O']
PLAY='YES'          
while PLAY=='YES':
  PLAYER=input('Which player starts the game(MAX/MIN):')
  if PLAYER=='MAX':
    MAX=1
    MIN=0
    SCORE=[4, 4, -3, 4, 6, -3, 7, -1, 4, 2, 6, -3, -5, 0, 7]
  elif PLAYER=='MIN':
    MAX=0
    MIN=1 
    SCORE=[0, 2, 0, -1, 2, -5, 0, -1, 4, 2, 6, -3, -5, 0, 7]    
  lastlevel=2
  N=0    
  currentmove='A' 
  currentscore=4  
  GAME_PATH=['A'] 
  print("STARTING STATE OF THE GAME IS", GAME_PATH)
  while N<=lastlevel:                                 
    if MAX==0 and MIN==1:                    
      N=N+1                                   
      print("MOVE", N)
      print("It's MIN's turn!")
      print ("What are the possible moves?")
      move1=input()                          
      move2=input()
      move1_pos=GAMETREE.index(move1)       
      move2_pos=GAMETREE.index(move2)
      if SCORE[move1_pos]<SCORE[move2_pos]:   
         print(move1, "selected")           
         currentmove=move1               
         currentscore = SCORE[move1_pos]     
      else:
         print(move2, "selected")          
         currentmove=move2                  
         currentscore = SCORE[move2_pos]     
      GAME_PATH.append(currentmove)         
      MAX=1                               
      MIN=0                              
    elif MAX==1 and MIN==0:                   
      N=N+1
      print("MOVE", N)
      print("It's MAX's turn!")
      print ("What are the possible moves?")
      move1=input()                         
      move2=input()
      move1_pos=GAMETREE.index(move1)       
      move2_pos=GAMETREE.index(move2)
      if SCORE[move1_pos]>SCORE[move2_pos]:  
         print(move1, "selected")           
         currentmove=move1                   
         currentscore = SCORE[move1_pos]     
      else:
         print(move2, "selected")          
         currentmove=move2                 
         currentscore = SCORE[move2_pos]     
      GAME_PATH.append(currentmove)          
      MAX=0                                   
      MIN=1                               
  if MAX==1:                                                
     print("The final score of MIN player is",currentscore)   
     print("The Game path is", GAME_PATH)
  else:                                                    
     print("The final score of MAX player is",currentscore)   
     print("The Game path is", GAME_PATH)
  print("\n \n")   
  CONTINUE=input('Do you want to continue(YES/NO):')    
  PLAY=CONTINUE  
```
