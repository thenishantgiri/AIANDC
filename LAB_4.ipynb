```python
# Program to create a Simple Calculator. 
# Write a menu driven program. Ask the user to enter the choice of "mathematical operation to be performed (+, -, * or /)"
# create a separate function for each arithmetic operation (addiiton, subtraction, multiplication and division)
# call the required function to carry out the mathematical operation as desired by the user

def add(x="0",y="0"):
  z=x+y
  return z
  
def sub(x="0",y="0"):
  if x>y:
    z=x-y
  else:
    z=y-x
  return z  

def mul(x="1",y="1"):
  z=x*y
  return z

def divide(x="1",y="1"):
  z=x/y
  return z  

print("Please enter your choice \n1. ADDITION\n2. SUBTRACTION\n3. MULTIPLICATION\n4. DIVISION\n ")
choice=int(input())   

print("Please enter two numbers")
a=int(input())
b=int(input())

if choice==1:
  c=add(a,b)
  print("Sum of", a,"and",b, "is",c)
elif choice==2:
  c=sub(a,b)
  print("Difference of", a,"and",b, "is",c)  
elif choice==3:
  c=mul(a,b)
  print("Product of", a,"and",b, "is",c) 
else:
  c=divide(a,b)
  print("Division of", a,"and",b, "is",c)   
```

```python
# PROGRAM TO CREATE A LIST CONTAINING NAMES OF TEN STUDENTS. 
# DISPLAY A MENU OPTION TO THE USER TO 
#1.ADD A NAME IN A LIST 2.REMOVE A NAME FROM THE LIST 3.SORT THE LIST IN ALPHABETICAL ORDER 4.SEARCH FOR A NAME IN A LIST
# CREATE A SEPARATE FUNCTION TO PERFORM ALL THESE OPERATIONS

def add(a):
  list.append(a)
  return list

def remove(a):
  list.remove(a)
  return list

def arrange(list):
  list.sort(reverse=False) 
  return list

def search(a):
  if a in list:
    print(a,"is in the list")
  else :
    print(a,"is not in the list")  

def Display(list):
  for x in list:
    print(x)

students=["Mayank","Kalayani","Akankasha","Deeptarun","Shubham","Seema","Manas","Nishant","Riya","Sreedevi"]
Display(students)
print("\n\n")
list=students

print("Please enter your choice \n1.ADD A NAME IN A LIST \n2.REMOVE A NAME FROM THE LIST \n3.SORT THE LIST IN ALPHABETICAL ORDER \n4.SEARCH FOR A NAME IN A LIST \n")
choice=int(input())  
if choice==3:
  arrange(students)
  print("NEW LIST IS:\n")
  Display(students)
else:
  a=input('Enter the name: ')
  if choice==1:
    add(a)
    print("NEW LIST IS:\n")
    Display(students)
  elif choice==2:
    remove(a)
    print("NEW LIST IS:\n")
    Display(students)
  elif choice==4:  
    search(a)
```
