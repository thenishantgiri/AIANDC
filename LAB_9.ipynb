```python
#TREE 1
PATH=[]     
s=0         
final='G'   
print("Please enter the elements of the tree\n")  
print("Please enter first element (root node)\n") 
root = input()      
PATH.append(root)   
if root==final:            
  print("Goal Node Found!") 
  print("The path is", root)
  s=1                   
currentnode = root            
ans='y'
while s==0 and (ans=='y'or ans=='Y'):            
  print ("Please enter the LEFT child of", currentnode)             
  left=input()
  print ("Please enter the path cost from",currentnode , "to", left) 
  GL=int(input())
  print ("Please enter the heuristic value of", left)               
  HL=int(input())
  print ("Please enter the RIGHT child of", currentnode)              
  right=input()
  print ("Please enter the path cost from",currentnode , "to", right) 
  GR=int(input())
  print ("Please enter the heuristic value of", right)                
  HR=int(input())
  FL = GL+HL        
  FR = GR+HR         
  if FL<FR:           
    PATH.append(left)   
    currentnode=left   
    if left==final:            
      print("Goal Node Found!") 
      print("The path is")
      print(PATH)              
      s=1                      
      break
  else:                       
    PATH.append(right)        
    currentnode=right       
    if right==final:          
      print("Goal Node Found!")  
      print("The path is")
      print(PATH)               
      s=1                       
      break
  print("PATH =", PATH)       
  print("Do you want to continue(y/n)?") 
  ans=input()
if s==0:
   print("Goal not found !")
```
