```python
#PROGRAM 1,2 and 3 on search tree 1
Tree=[]   
i=0
j=0
print("Please enter element ", i, "(root node)")  
a=input() 
Tree.append(a)      
i=i+1              
while i<7:
    print("Please enter RIGHT child of element of", Tree[j])
    print("If there is no RIGHT child , enter 'null'")
    a=input() 
    if a!='null':
      Tree.append(a)       
      i=i+1
    print("Please enter LEFT child of element ", Tree[j])
    print("If there is no LEFT child , enter 'null'")
    a=input() 
    if a!='null':
      Tree.append(a)
      i=i+1
    j=j+1 
def BFS(Tree):
  Queue=[ ]                        
  initial =input('Please enter the initial state: ')                       
  pos= Tree.index(initial)                
  goal= input('Pleas enter the goal state: ')
  Queue.append(initial)               
  s=0                              
  if Queue[0]==goal:                                   
     print("Goal Node Found ! Search Successful!")
     print(Queue[0],"is the goal node")                    
     s=1
  if s==0:                 
     print(Queue[0])           
     Queue.pop(0)             
     k=pos+1  
     while k<=6:              
       Queue.append(Tree[k])    
       k=k+1                  
       if k>6:
         break
       Queue.append(Tree[k])
       k=k+1
       if k>6:
         break
       print(Queue[0])                               
       if Queue[0]==goal:                             
        print("Search Successful! Goal Node Found!")   
        s=1                                         
        break                                     
       Queue.pop(0)                                 
  if s==0:                
    while Queue!=[]:
       print(Queue[0])                                 
       if Queue[0]==goal:                                
          print("Search Successful! Goal Node Found !")   
          s=1                                         
          break
       Queue.pop(0)                              
  if s==0: 
    print("Goal Node not found! Search not successful !")    
e=1
while e<=3:
  print("PROGRAM",e)
  BFS(Tree)
  e+=1
```
